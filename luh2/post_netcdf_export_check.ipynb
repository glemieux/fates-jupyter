```python
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
def plotnatpft(variable, colormap='magma_r'):
    fg = variable.plot(x='lon',y='lat',col='natpft', col_wrap=4, cmap=colormap, 
                   transform=ccrs.PlateCarree(), subplot_kws=dict(projection=ccrs.Robinson()), add_colorbar=False)

    for ax in fg.axes.flat:
        ax.coastlines() 

    # color bar
    fg.add_colorbar(orientation="vertical", shrink=0.6, anchor=(1.5, 0.5))
    plt.subplots_adjust(hspace=-0.65)
```

```python
ds_netcdf = xr.open_dataset("/home/glemieux/Repos/fates/tools/landuse/src/landusepft/fates_landuse_pft_map_4x5.nc")
```

### Bareground

```python
ax = plt.subplot(projection=ccrs.Robinson())
ds_netcdf.pct_brgnd.plot(cmap='magma_r', transform=ccrs.PlateCarree())
ax.coastlines()
```

### Current surface

```python
plotnatpft(ds_netcdf.pct_csurf)
```

### Primary

```python
plotnatpft(ds_netcdf.pct_primr)
```

### Pasture

```python
plotnatpft(ds_netcdf.pct_pastr)
```

### Range

```python
plotnatpft(ds_netcdf.pct_range)
```

### Potential

```python
potential_minus_present = ds_netcdf.pct_primr - ds_netcdf.pct_csurf
```

```python
plotnatpft(potential_minus_present,colormap='PRGn')
```

# Brazil surface data set monthly averages

This notebook simply reports the values of the monthly averages for the surface data set that is used for the `1x1_brazil` resolution.

## Setup notebook

```julia
# https://alexander-barth.github.io/NCDatasets.jl/stable/
using NCDatasets
using LinearAlgebra
using Statistics
```

## Import data

```julia
surfdata = NCDataset("surfdata_1x1_brazil_hist_16pfts_Irrig_CMIP6_simyr2000_c190214.nc");
```

```julia
surfdata.dim
```

```julia
sai = surfdata["MONTHLY_SAI"];
```

```julia
sai.attrib
```

```julia
# Yearly mean for each pft
saidat = sai[1,1,:,:]
```

```julia
# Monthly sai value for a given pft
pftnum = 2
saidat[pftnum,:]
```

```julia
# Monthly mean for all pfts
mean(saidat,dims=1)
```

```julia
# Yearly mean for each PFT
mean(saidat,dims=2)
```

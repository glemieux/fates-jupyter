## Import libraries

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import ctsm_py.fates_xarray_funcs as fx
```

## Load history output data

```python
fn_base = 'ERP_Ld3.f09_g17.I2000Clm50Fates.cheyenne_intel.clm-FatesColdDef.C.spmode-baseline-diagout.clm2.h0.2000-01-01-00000.nc'
fn_rest = 'ERP_Ld3.f09_g17.I2000Clm50Fates.cheyenne_intel.clm-FatesColdDef.C.spmode-postmerge-api16-1-diagout.clm2.h0.2000-01-01-00000.nc'
data_base = xr.open_dataset(fn_base)
data_rest = xr.open_dataset(fn_rest)
```

## Check the dimension mapping size

```python
data_base.TLAI.dims
```

```python
data_base.TLAI.time.size
```

## Plot the data

### TLAI

```python
tlai_base = data_base.TLAI
tlai_rest = data_rest.TLAI
tlai_diff = tlai_base - tlai_rest
```

```python
tstop_idx = 0
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 4))
tlai_base.isel(time=tstop_idx).plot(ax=ax1)
tlai_rest.isel(time=tstop_idx).plot(ax=ax2)
tlai_diff.isel(time=tstop_idx).plot(ax=ax3)
```

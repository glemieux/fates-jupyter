# Fixed Biogeography small pft area comparison

This simple notebook plots out the `PFTbiomass` and areas.  The original sp mode code was removing small pft areas and causing non-b4b changes.

## Import libraries

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import ctsm_py.fates_xarray_funcs as fx
```

## Load history output data

```python
fn_base = 'ERS_Ld3.f45_f45_mg37.I2000Clm50FatesCru.cheyenne_intel.clm-FatesReducedComplexFixedBiogeo.spmode-baseline-fixedbiogeo.clm2.h0.2000-01-01-00000.nc'
fn_rest = 'ERS_Ld3.f45_f45_mg37.I2000Clm50FatesCru.cheyenne_intel.clm-FatesReducedComplexFixedBiogeo.C.spmode-shortfbg-check.clm2.h0.2000-01-01-00000.nc'
data_base = xr.open_dataset(fn_base)
data_rest = xr.open_dataset(fn_rest)
```

## Plot the data

### PFTbiomass

```python
levpft = 0
pftbm_base = data_base.PFTbiomass.isel(fates_levpft=levpft)
pftbm_rest = data_rest.PFTbiomass.isel(fates_levpft=levpft)
pftbm_diff = pftbm_base - pftbm_rest
```

```python
tstop_idx = 0
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 4))
pftbm_base.isel(time=tstop_idx).plot(ax=ax1)
pftbm_rest.isel(time=tstop_idx).plot(ax=ax2)
pftbm_diff.isel(time=tstop_idx).plot(ax=ax3)
```

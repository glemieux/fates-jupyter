## Plot the data

### Secondary forest fraction

Variable diff calculations

```python
sff_elm_base = elm_base.SECONDARY_FOREST_FRACTION
sff_elm_rest = elm_rest.SECONDARY_FOREST_FRACTION
sff_elm_diff = sff_elm_base - sff_elm_rest
```

Plot new `all_soil_patches` loop base-rest difference

```python
tstop_idx = 30
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 4))
sff_elm_base.isel(time=tstop_idx).plot(ax=ax1)
sff_elm_rest.isel(time=tstop_idx).plot(ax=ax2)
sff_elm_diff.isel(time=tstop_idx).plot(ax=ax3)
```

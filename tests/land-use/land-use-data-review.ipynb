```python
fn_cd_base = 'ERS_D_Ld30.f45_f45_mg37.I2000Clm50FatesCru.cheyenne_intel.clm-FatesColdDef.testupdate-FatesColdDef.clm2.h0.2000-01-01-00000.nc.base'
fn_cd_rest = 'ERS_D_Ld30.f45_f45_mg37.I2000Clm50FatesCru.cheyenne_intel.clm-FatesColdDef.testupdate-FatesColdDef.clm2.h0.2000-01-01-00000.nc.rest'

cd_base = xr.open_dataset(fn_cd_base)
cd_rest = xr.open_dataset(fn_cd_rest)
```

## Plot the data

### Secondary forest fraction

Variable diff calculations

```python
sff_lu_base = lu_base.SECONDARY_FOREST_FRACTION
sff_lu_rest = lu_rest.SECONDARY_FOREST_FRACTION
sff_lu_diff = sff_lu_base - sff_lu_rest

sff_cd_base = cd_base.SECONDARY_FOREST_FRACTION
sff_cd_rest = cd_rest.SECONDARY_FOREST_FRACTION
sff_cd_diff = sff_cd_base - sff_cd_rest

sff_diff = sff_cd_base - sff_lu_base
```

Plot the difference between the LandUse testmod and standard cold start run

```python
tstop_idx = 30
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 4))
sff_cd_base.isel(time=tstop_idx).plot(ax=ax1)
sff_lu_base.isel(time=tstop_idx).plot(ax=ax2)
sff_diff.isel(time=tstop_idx).plot(ax=ax3)
```

### Secondary forest biomass

Variable diff calculations

```python
sfb_lu_base = lu_base.SECONDARY_FOREST_BIOMASS
sfb_lu_rest = lu_rest.SECONDARY_FOREST_BIOMASS
sfb_lu_diff = sfb_lu_base - sfb_lu_rest

sfb_cd_base = cd_base.SECONDARY_FOREST_BIOMASS
sfb_cd_rest = cd_rest.SECONDARY_FOREST_BIOMASS
sfb_cd_diff = sfb_cd_base - sfb_cd_rest

sfb_diff = sfb_cd_base - sfb_lu_base
```

```python
tstop_idx = 30
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 4))
sfb_cd_base.isel(time=tstop_idx).plot(ax=ax1)
sfb_lu_rest.isel(time=tstop_idx).plot(ax=ax2)
sfb_diff.isel(time=tstop_idx).plot(ax=ax3)
```
